```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
total_customers = 1000000
```

```python
tenures = ([1]*200000 + [2]*300000 + [3]*400000 + [4]*100000)
```

```python
np.random.shuffle(tenures)
```

```python
dates= pd.date_range("2024-01-01", "2024-12-31")
purchase_dates= np.random.choice(dates, total_customers)
```

```python
df= pd.DataFrame({
    "Customer_ID":range(1, total_customers+1),
    "Vehicle_ID": range(1, total_customers+1),
    "Vehicle_Value": 100000,
    "Policy_Purchase_Date": purchase_dates,
    "Policy_Tenure": tenures
})
```

```python
df["Premium"]= df["Policy_Tenure"]*100
df["Policy_Start_Date"]= df["Policy_Purchase_Date"] + pd.Timedelta(days=365)
df["Policy_End_Date"]= df["Policy_Start_Date"] + pd.to_timedelta(df["Policy_Tenure"]*365, unit="d")
```

```python
df.to_csv("policy_sales_data.csv", index=False)
```

```python
claims= []
claim_dates= [7,14,21,28]
```

```python
eligible= df[df["Policy_Purchase_Date"].dt.day.isin(claim_dates)]
```

```python
sample= eligible.sample(frac= 0.30)
```

```python
for i,row in sample.iterrows():
    if row["Policy_Start_Date"]<= row["Policy_End_Date"]:
        claims.append({
            "Claim_ID": len(claims)+1,
            "Customer_ID": row["Customer_ID"],
            "Vehicle_ID": row["Vehicle_ID"],
            "Claim_Amount": row["Vehicle_Value"]*0.10,
            "Claim_Date": row["Policy_Start_Date"],
            "Claim_Type": 1
        })

claims_df= pd.DataFrame(claims)
```

```python
claims_df.to_csv("claims_data.csv", index=False)
```
